```python
## Problem 2-1
def equgal(ra, dec):
    sinb = (np.cos(1.093)*np.sin(dec)) - (np.sin(1.093)*np.cos(dec)*np.sin(ra-4.926))
    b = np.arcsin(sinb)
    cosl = np.cos(dec)*np.cos(ra-4.926)/np.cos(b)
    l = np.arccos(cosl)+0.576
    return np.degrees(l), np.degrees(b)
```

```python
## Problem 2-2
l, b = equgal(np.radians(data['ra']),np.radians(data['dec']))
fig, ax = plt.subplots(figsize=(12,8))
fig.gca().invert_xaxis()

ax.scatter(l, b,marker='.',s=0.1)
ax.set_title('Galactic Longitude l vs Galactic Latitude b')
ax.set_xlabel('l (deg)')
ax.set_ylabel('b (deg)')
fig.savefig('problem2-2.pdf')
```

```python
data2 = np.genfromtxt('../small_sdss.csv', delimiter=',', names=True, unpack=False)
```

```python
## Problem 3-1
def gc_dist(lon1, lat1, lon2, lat2):
    cosdist = (np.sin(lat1)*np.sin(lat2)) + (np.cos(lat1)*np.cos(lat2)*np.cos(np.abs(lon1-lon2)))
    return np.degrees(np.arccos(cosdist))
```

```python
## Problem 3-2
fig, ax = plt.subplots(figsize=(12,8))

ax.scatter(data2['ra'],data2['dec'],marker='.',s=0.5)
ax.set_title('SDSS Right Ascension vs Declination')
ax.set_xlabel('RA (deg)')
ax.set_ylabel('Dec (deg)')
fig.savefig('problem3-2.pdf')
```

```python
## Problem 3-3
ind = np.triu_indices(len(data2), 1)
ra1 = np.radians(data2['ra'][ind[0]])
ra2 = np.radians(data2['ra'][ind[1]])
dec1 = np.radians(data2['dec'][ind[0]])
dec2 = np.radians(data2['dec'][ind[1]])

dist = gc_dist(ra1, dec1, ra2, dec2)
```

```python
## Problem 3-4
fig, ax = plt.subplots(figsize=(12,8))

ax.hist(dist,bins=100)
ax.set_title('Spherical Distance Histogram')
ax.set_xlabel('Angular Separation (deg)')

fig.savefig('problem3-4.pdf')
```

```python
## Problem 3-5

### Create a 2-vector of 1000 random and uniformly distributed floats in the interval [0,1]
randobj = np.random.rand(2,2000)
### Calculate RA and Dec
dec = np.arccos((2*randobj[0])-1)-(np.pi/2.)
ra = 2*np.pi*randobj[1]
### Create an index array for finding distances between objects using triu_indices
ind = np.triu_indices(len(randobj[0]), 1)

ra1 = ra[ind[0]]
ra2 = ra[ind[1]]
dec1 = dec[ind[0]]
dec2 = dec[ind[1]]
dist = gc_dist(ra1, dec1, ra2, dec2)
```

```python
fig, ax = plt.subplots(figsize=(12,8))

ax.hist(dist,bins=100)
ax.set_title('Spherical Distance Histogram')
ax.set_xlabel('Angular Separation (deg)')

fig.savefig('problem3-5.pdf')

### This is roughly a Poisson distribution
```
